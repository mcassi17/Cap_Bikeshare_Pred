```python
import pandas as pd
```

```python
def dictionary(row):
    stat = row['start_station']
    distance = row['metro_dist']
    dict_map[stat] = distance
```

```python
rush = pd.read_csv('rush_hour.csv')
rush = rush.drop('Unnamed: 0', 1)
rush.head()
```

```python
metro = pd.read_csv('station_distances.csv')
```

```python
dict_map = {}
metro.apply(dictionary, axis=1);
```
