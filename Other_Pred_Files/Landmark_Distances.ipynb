```python
lat_lon = [(38.889931, -77.009003),
          (38.897957, -77.036560),
          (38.889484, -77.035278),
          (38.8978218, -77.00609179999998),
          (38.879074, -77.069006),
          (38.906975, -77.065081),
          (38.872021, -77.00885),
          (38.8998, -77.0217),
          (38.909649, -77.043442),
          (38.8903896, -77.0696999),
          (38.88963, -77.022979),
          (38.881412, -77.036476),
          (38.889279, -77.050188)]
landmark = ['capitol', 
           'whitehouse',
            'wash_mon',
           'union', 
           'arlington_cem', 'georgetown', 'natspark', 'chinatown','dupont', 'iwojima',
           'natmall', 'jefferson', 'lincoln']
len(landmark),len(lat_lon)
```

```python
dictionary = {'landmark': landmark,
             'lat_lon': lat_lon}
landmarks = pd.DataFrame(dictionary)
```

```python
start = pd.read_csv('/Users/matthewcassi/Documents/Bike-Sharing-Dataset/Bikeshare_Time_Prediction/metro_rush.csv', 
                        parse_dates=['Start date', 'End date', 'start_date_short', 'end_date_short'])
start.drop('Unnamed: 0', 1, inplace=True)
include = ['start_station','start_lat','start_lon']
start = start[include]
```

```python
start['lat_lon'] = list(zip(start.start_lat, start.start_lon))
start = start.drop(['start_lat', 'start_lon'], 1)
start = start.drop_duplicates()
```

```python
landmark_lats = list(landmarks['lat_lon'])
```

```python
def landmark_dist(row):
    distances = []
    for i in landmark_lats:
        distance = vincenty(row['lat_lon'], i).miles
        distances.append(distance)
    min_d = min(distances)
    return min_d
```

```python
start['landmark_dist'] = start.apply(landmark_dist, axis=1)
```

```python
start.reset_index().drop('index', 1, inplace=True)
```
